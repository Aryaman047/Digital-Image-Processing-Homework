```python
from matplotlib import pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt
from skimage.color import rgb2gray
import glob
from skimage.io import imread
```

```python
example_file = glob.glob(r"C:\Users\aryam\Downloads\HD-wallpaper-stars-in-sky-3-abstract-in-sky-stars.jpg")[0]
im = imread(example_file)
plt.imshow(im)
plt.show()
```

```python
blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.1)
# Compute radii in the 3rd column.
blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
numrows = len(blobs_log)
print("Number of stars counted : " ,numrows)
```
